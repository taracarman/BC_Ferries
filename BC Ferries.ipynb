```python
capacity = pd.pivot_table(ferries, index="Route", columns="Over_Capacity", values="Sailing_Time", aggfunc=np.count_nonzero)
```

```python
capacity["Rate_Ovr_Cap"] = capacity["Yes"]/capacity["Total"]
```

## Worst waits

Which routes are most likely to have more than 100 cars waiting.

```python
more_100 = pd.pivot_table(ferries, index="Route", columns="More_than_100", values="Sailing_Time", aggfunc=np.count_nonzero)
```

```python
more_100.head()
```

```python
more_100["Total"] = more_100.sum(axis=1)
```

```python
more_100["Rate_More_100"] = more_100["Yes"]/more_100["Total"]
```

For sailings that left more than 100 cars behind, Swartz Bay-Tsawwassen still had the highest rate, at 15 per cent, though it was followed closely by Horseshoe Bay-Departure Bay.

```python
more_100.head()
```

```python
more_500 = pd.pivot_table(ferries, index="Route", columns="More_than_500", values="Sailing_Time", aggfunc=np.count_nonzero)
```

```python
more_500["Total"] = more_500.sum(axis=1)
```

```python
more_500["Rate_More_500"] = more_500["Yes"]/more_500["Total"]
```

There were 23 sailings on the Departure Bay-Horseshoe Bay route that left more than 500 cars behind, and 11 on the Swartz Bay to Tsawwassen.

```python
more_500.head()
```

The really bad waits were almost all on the Departure Bay side (headed to Horseshoe Bay) and the Swartz Bay side (headed to Tsawwassen). 
The worst times were the afternoon of Victoria Day and Sunday afternoons in the summer.

```python
ferries[ferries.More_than_500 == "Yes"]
```

## Delayed or cancelled sailings

```python
delay = pd.pivot_table(ferries, index ="Route", columns="Delay_Reason", values="Sailing_Time", aggfunc=np.count_nonzero)
delay.head()
```

Traffic was the reason the most sailings were delayed. 

```python
delay["Total"] = delay.sum(axis=1)
delay.head()
```

Horseshoe Bay routes were the most likely to experience delays. Almost one in three (31%) sailings between Horseshoe Bay and Langdale and one in five (24%) between Horseshoe Bay and Departure Bay were delayed in 2016. 

```python
ferries.Delay_Reason.value_counts()
```

```python
mechanics = ferries[(ferries.Delay_Reason == "M-Accumulated vessel mechanical delays") | (ferries.Delay_Reason == "M-Vessel-other problems") | (ferries.Delay_Reason == "M-Vessel propulsion system")]
```

The Spirit vessels had the most delays due to mechanical reasons.

```python
mechanics.Vessel.value_counts()
```

```python
cancel = pd.pivot_table(ferries, index ="Route", columns="Cancel_Reason", values="Sailing_Time", aggfunc=np.count_nonzero)
cancel.head()
```

```python
cancel["Total"] = cancel.sum(axis=1)
cancel.head()
```
